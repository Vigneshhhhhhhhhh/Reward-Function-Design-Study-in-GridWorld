## 4. Key Findings Summary
* **Dense vs Sparse**: Dense rewards often converge faster.
* **Negative Step Penalty**: Highly effective for shortest-path optimization.
* **Deceptive Rewards**: The agent often gets 'stuck' in the fake reward square, proving how fragile RL can be when designers give misleading incentives.
