## 2. Introduction
This study investigates how different **Reward Functions** (the 'treats' for the AI) affect the learning speed and efficiency of a Reinforcement Learning agent in a stochastic **GridWorld** environment.

### Why is this important?
Reinforcement Learning (RL) agents learn by trial and error based on a reward signal. If the signal is too sparse (only at the end), the agent gets lost. If it's too frequent but wrong, the agent learns bad behavior. This project compares 5 distinct designs.

## 3. The 5 Reward Structures Compared
1. **Dense Reward**: Small positive rewards for every step towards the goal.
2. **Sparse Reward**: Only one reward given at the terminal goal state.
3. **Shaped Reward**: A potential-based reward that guides the agent via distance-to-goal.
4. **Negative Step Penalty**: A small penalty for every move, forcing the agent to find the shortest path.
5. **Deceptive Reward**: A fake local maximum that 'tricks' the agent away from the true goal.

## 6. Key Findings Summary
Based on the data generated in the `results/findings.txt`:
* **Dense vs Sparse**: Dense rewards often converge faster but can lead to suboptimal 'sub-looping' if not carefully designed.
* **Negative Step Penalty**: Highly effective for shortest-path optimization.
* **Deceptive Rewards**: The agent often gets 'stuck' in the fake reward square, proving how fragile RL can be when designers give misleading incentives.
